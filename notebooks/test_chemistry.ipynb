```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
import prodimopy.read as pread
import prodimopy.plot as pplot
```

```python
model=pread.read_prodimo()
```

```python
pp=pplot.Plot(None)
```

```python
chemistry=pread.analyse_chemistry('CO',model)
```

```python
pp.plot_cont(model,chemistry.farray);
```

```python
k=0
plt.figure(figsize=(model.nx/3,model.nz/3))
for i in range(1,model.nx+1):
    for j in range(1,model.nz+1)[::-1]:
        plt.text(i-1, j-1, chemistry.main_form_grid[k])
        k+=1
plt.imshow(np.log10(chemistry.farray.T))
plt.xlim(0,model.nx)
plt.ylim(0,model.nz)
```
